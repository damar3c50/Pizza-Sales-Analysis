```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv('/content/drive/MyDrive/Datasets/pizza_sales (3).csv')
df.head()
```

```python
df.describe()
```

```python
df.info()
```

```python
#Drop kolom unnamed:0
df.drop(columns=['Unnamed: 0'], inplace=True)
df.head()
```

```python
#Ubah tipe data pada kolom order_date
df['order_date'] = pd.to_datetime(df['order_date'], dayfirst=True)
df.info()
```

```python
df.head()
```

```python
x = df['pizza_name'].value_counts().reset_index()
fig, ax = plt.subplots(figsize=(7, 10))
ax.barh(x['pizza_name'], x['count'])
```

```python
df['month'] = df['order_date'].dt.month
df.head()
```

```python
df['week'] = df['order_date'].dt.isocalendar().week
df
```

```python
df['formatted_time'] = pd.to_datetime(df['order_time']).dt.strftime('%H:00')
```

```python
df.duplicated().sum()
```

```python
df.isnull().sum()
```

```python
#Profit berasal dari 20% harga setiap pizza
df['profit_per_product'] = df['unit_price'] * 0.2
df.head()
```

```python
#Mencari pendapatan dan keuntungan (harian, mingguan, bulanan) ***
df['revenue_daily'] = df.groupby('order_date')['total_price'].transform('sum')
df['profit_daily'] = df.groupby('order_date')['profit_per_product'].transform('sum')
df['revenue_weekly'] = df.groupby('week')['total_price'].transform('sum')
df['profit_weekly'] = df.groupby('week')['profit_per_product'].transform('sum')
df['revenue_monthly'] = df.groupby('month')['total_price'].transform('sum')
df['profit_monthly'] = df.groupby('month')['profit_per_product'].transform('sum')
df.head()
```

```python
#Daily Revenue
x = df[['order_date','revenue_daily']].drop_duplicates()
fig = plt.subplots(figsize=(15, 3))
sns.lineplot(data=x, x='order_date', y='revenue_daily')
plt.title('Daily Revenue')
plt.xlabel('Date')
plt.ylabel('Revenue')
```

```python
#Daily Profit
x = df[['order_date','profit_daily']].drop_duplicates()
fig = plt.subplots(figsize=(15, 3))
sns.lineplot(data=x, x='order_date', y='profit_daily')
plt.title('Daily Profit')
plt.xlabel('Date')
plt.ylabel('Profit')
```

```python
#Weekly Revenue
x = df[['week','revenue_weekly']].drop_duplicates()
fig = plt.subplots(figsize=(15, 3))
sns.lineplot(data=x, x='week', y='revenue_weekly')
plt.title('Weekly Revenue')
plt.xlabel('Week')
plt.ylabel('Revenue')
```

```python
#Weekly Profit
x = df[['week','profit_weekly']].drop_duplicates()
fig = plt.subplots(figsize=(15, 3))
sns.lineplot(data=x, x='week', y='profit_weekly')
plt.title('Weekly Profit')
plt.xlabel('Week')
plt.ylabel('Profit')
```

```python
#Monthly Revenue
x = df[['month','revenue_monthly']].drop_duplicates()
fig = plt.subplots(figsize=(15, 3))
sns.lineplot(data=x, x='month', y='revenue_monthly')
plt.title('Monthly Revenue')
plt.xlabel('Month')
plt.ylabel('Revenue')
```

```python
#Monthly Profit
x = df[['month','profit_monthly']].drop_duplicates()
fig = plt.subplots(figsize=(15, 3))
sns.lineplot(data=x, x='month', y='profit_monthly')
plt.title('Monthly Profit')
plt.xlabel('Month')
plt.ylabel('Profit')
```

```python
#Menghitung pendapatan dari masing-masing produk (harian, mingguan, bulanan)
df['total_revenue_per_product_daily'] = df.groupby(['order_date', 'pizza_name'])['total_price'].transform('sum')
df['total_revenue_per_product_weekly'] = df.groupby(['week', 'pizza_name'])['total_price'].transform('sum')
df['total_revenue_per_product_monthly'] = df.groupby(['month', 'pizza_name'])['total_price'].transform('sum')
df.head()
```
